```python
# importación de pandas
import pandas as pd 

# importación de matplotlib. 
from matplotlib import pyplot as plt
```

```python
# Ejemplo sencillo de uso de pyplot
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]
# con la función plot() ploteamos la variable x en el eje x y la variable y en el eje y.
plt.plot(x, y)
# también podemos crear un segundo plot
plt.plot(x, z)
# Con esta instrucción añadimos un título
plt.title("First plot")
# Con esta instrucción añadimos un nombre al eje x
plt.xlabel("X")
# Con esta instrucción añadimos un nombre al eje y
plt.ylabel("Y and Z")
# Podemos añadir una leyenda con la función legend(). Esta función recibe como argumento una lista que dará una leyenda para cada plot
plt.legend(["X vs Y", "X vs Z"])
```
